Problem 4 Matrix Report


```python
import numpy as np

def generate_and_solve():
    results = []
    count = 0

    while count < 3:
        #creating random 3x3 matrix with certain range
        M = np.random.randint(-10, 11, size=(3, 3))

        #making sure the matrix is invertible
        if np.linalg.det(M) != 0:
            #inverse computation
            M_inv = np.linalg.inv(M)

            #matrix * inverse
            product = np.dot(M, M_inv)

            results.append((M, M_inv, product))
            count += 1

    return results

matrices_data = generate_and_solve()
matrices_data
```

```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')

#excluding 'id', 'date', and 'zipcode'.
columns_to_drop = ['id', 'date', 'zipcode']
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')

#outlining the dependant variable and features
response_col = 'price'
features = df_cleaned.drop(columns=[response_col])
response = df_cleaned[response_col]

#question 1, computing average, min, max, and variance
print("Question 1: Feature Statistics")

#computing stats for each column and putting them into a df
stats_df = pd.DataFrame()
stats_df['Average'] = features.mean()
stats_df['Min'] = features.min()
stats_df['Max'] = features.max()
stats_df['Variance'] = features.var()

print(stats_df)
print("\n")

#question 2, correlation coefficients of all features with price
print("Question 2: Correlation with Price")

#calculating correlation
correlations = features.corrwith(response)

#making it a df for table like display
correlation_table = pd.DataFrame(correlations, columns=['Correlation Coefficient'])

print(correlation_table)
```
